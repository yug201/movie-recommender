```python
movies=pd.read_csv(r"C:\Users\Yug\OneDrive\Desktop\Movi recomandation system\tmdb_5000_movies.csv")
credits=pd.read_csv(r"C:\Users\Yug\OneDrive\Desktop\Movi recomandation system\tmdb_5000_credits.csv")
```

```python
# mearging  datasets on basis of  titel of movies
movies=movies.merge(credits,on="title")
```

# Data Processing

```python
#removing  unnescessory  features  
movies = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
print(movies.isnull().sum(),"\n duplicate  values \n", movies.duplicated().sum())
# onley overvies  have  some missing  values  since   onley  3  null values are present   so we will simpley   drom  thes   rows
movies.dropna(inplace=True)
```

```python
#  geners ,keywords feature  values  are in dictonarey format so we have to  change it 
# print(movies.iloc[0].genres)
import ast
def convert(itm):
    l=[]
    for x in ast.literal_eval(itm): 
        l.append(x["name"])
    return l
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
        
```

```python
def convert3(text):
    L = []
    for i in ast.literal_eval(text):
        if len(L)<3:
            L.append(i['name'])
    return L 
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
#  we  did not need to  consider  entire  crew   information
ast.literal_eval(movies["crew"][0])
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new["tag"]=new["tags"].apply(lambda x:x.lower())
new
```

```python
# Text → Stemming → Vectorization

import snowballstemmer
stemmer=snowballstemmer.stemmer('english')
def stem_text(text):
    return " ".join(stemmer.stemWords(text.split()))
# Apply on dataframe FIRST
new['tags']=new['tags'].apply(stem_text)

# THEN vectorize
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vector=cv.fit_transform(new['tags']).toarray()
```

# cosine_similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector) # this  will calculate the  similarity of  each  vector  with other vector  
similarity.shape  # this will contin 4806 array of   4806 len which is  distance of  each  movi  with  other 
```

```python
similarity[0] #  this  will be similarity of  movi one with  other  movies
```

```python
idx_titl={}
for i,x in enumerate(new["title"]):
    idx_titl[x]=i
```

```python
def recommend(movi,k=5):  #O(nlogn + k)==O(nlogn)  #  we can use heap in order to  descreas  time complexity  to  (O(n*log k))
    idx=idx_titl[movi]    #O(1)
    distances=similarity[idx]
    movies_list=sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])   #o(nlogn)
    res=[]
    for x in movies_list[1:min(k+1,len(idx_titl))]:  #O(k)
        res.append(new["title"][x[0]])
    return res
recommend("Avatar",5)
```
